```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num = np.array([16,	256,4096,65536,262144,1048576,4194304,16777216,67108864])
cpu = np.array([0.0004,0.0184,0.292,5.1027,19.2156,77.7799,308.531,1241.093,4954.43])
ompc = np.array([0.5128,0.5794,0.624,2.2955,7.4822,23.1049,89.3847,339.461,1333.735])
ompg = np.array([9.01211,8.87213,9.15733,9.15422,11.50514,14.65961,25.6641,64.804,222.944])
cuda = np.array([0.0196608,0.0171008,0.0175104,0.0654336,0.0690176,0.083968,0.260608,0.796877,2.88676])
```

```python
print(cpu/cuda)
print(cpu/ompg)
```

```python
plt.figure(dpi=150)
plt.plot(num, cpu/cuda,'o-',label="cuda")
plt.plot(num, cpu/ompc,'o-',label="openmp_CPU")
plt.plot(num, cpu/ompg,'o-',label="openmp_GPU")
plt.xscale('log')
plt.yscale('log')
plt.xlabel('file size (bytes)', fontweight ='bold', fontsize = 12)
plt.ylabel('Speed up', fontweight ='bold', fontsize = 12)
plt.title("AES speedup (log-log plot)")
plt.legend()
plt.show()
```

```python
plt.figure(dpi=150)
plt.plot(num, cuda,'o-',label="cuda")
plt.plot(num, ompc,'o-',label="openmp_CPU")
plt.plot(num, ompg,'o-',label="openmp_GPU")
plt.plot(num, cpu,'o-',label="sequential")
plt.xscale('log')
plt.yscale('log')
plt.xlabel('file size (bytes)', fontweight ='bold', fontsize = 12)
plt.ylabel('Time (ms)', fontweight ='bold', fontsize = 12)
plt.title("AES execution time (log-log plot)")
plt.legend()
plt.show()
```
